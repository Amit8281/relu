```python
# Create a WebDriver instance 
driver = webdriver.Chrome()

# Navigate to the initial URL
initial_url = "https://www.youtube.com/@tseries/videos"
driver.get(initial_url)

```

```python
# vedio_name = []
vedio_links = []

while True:
    
    
    try:
       
        print(f"Running loop number {i}")
        
#         # Scrape video names
#         name_tags = driver.find_elements(By.ID,"vote-count-middle")
#         vedio_name.extend([name.text for name in name_tags])
        
        # Scrape links
        vedio_link_tags = driver.find_elements(By.XPATH, '//a[@class="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media"]')
        vedio_links.extend([link.get_attribute("href") for link in vedio_link_tags])
        
        while True:
            height = driver.execute_script("return document.body.scrollHeight")
            driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
            
            if int(height) == 0:
                break
        
        time.sleep(2)  # Add the sleep after the while loop
        
    except NoSuchElementException:
        # Handle the exception if elements are not found on the current page
        break  # Exit the loop if needed
```
