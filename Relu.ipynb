```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains

from selenium.webdriver.chrome.options import Options     #importing Options
chrome_options = Options()
chrome_options.add_argument('--incognito')

import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.keys import Keys
```

## Scraping Url

```python
# vedio_url = []
vedio_links = []


# Create a WebDriver instance 
driver = webdriver.Chrome()

# Navigate to the initial URL
initial_url = "https://www.youtube.com/@tseries/videos"
driver.get(initial_url)

for i in range(1,30):
    try:
        
        # Scrape links
        vedio_link_tags = driver.find_elements(By.XPATH, '//a[@class="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media"]')
        vedio_links = [link.get_attribute("href") for link in vedio_link_tags]
        
        
        #Scroll Down
        while True:
            
            height = driver.execute_script("return document.body.scrollHeight")
            driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)

            if int(height) == 0:
                    break
            
        
        time.sleep(2)  
        
    except NoSuchElementException:
        break  
```

## Scraping Date By Url

```python
from selenium import webdriver


Date = []
# Create a WebDriver instance
driver = webdriver.Chrome()


# Loop through the URLs
for url in vedio_links[286:670]:
    try:
        # Navigate to the current URL
        driver.get(url)

        time.sleep(2)
        
        #Clicking on more
        more = driver.find_element(By.XPATH, '/html/body/ytd-app/div[1]/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[2]/ytd-watch-metadata/div/div[4]/div[1]/div/ytd-text-inline-expander/tp-yt-paper-button[1]')
        more.click()
        # Scrape data from the current page
        date = driver.find_elements(By.XPATH, '/html/body/ytd-app/div[1]/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[2]/ytd-watch-metadata/div/div[4]/div[1]/div/div[1]/yt-formatted-string/span[3]')
        Date.extend([name.text for name in date])

        

        # Return to the previous page (the list of URLs)
        driver.back()
        time.sleep(2)
    
    except NoSuchElementException:
        
        break  

```

## Counting Most Frequent Charactor From Url

```python
import pandas as pd
from collections import Counter
```

```python
# Extract characters after "=" in each URL and count them
def extract_and_count_chars(url):
    characters = url.split('=')[-1]
    return characters

data['characters'] = data['Songs_link'].apply(extract_and_count_chars)

# Concatenate all characters into a single string
all_characters = ''.join(data['characters'])

# Count the most frequent character
character_counts = Counter(all_characters)
most_common_character, count = character_counts.most_common(1)[0]

# Output the result
result = {
    'Key': most_common_character,
    'Value': count
}

print(result)
```
